```python
import matplotlib.pyplot as plt
#Line plot
days=[1,2,3,4,5]
sales=[100,200,300,400,500]
plt.plot(days,sales)
plt.title("Days vs sales")
plt.xlabel("day")
plt.ylabel("Sale")
plt.show()
```

```python
#Bar chart
std=['a','b','c','d']
marks=[85,95,45,65]
plt.plot(std,marks)
plt.xlabel("std")
plt.ylabel("marks")
plt.show()
```

```python
#Histogram
#ages
import numpy as np
ages=np.random.randint(20,30,size=10)
plt.hist(ages)
plt.show()
```

```python
# #Scatter plot
# income vs loan
inc= [1000,2000,3000,4000,5000]
loan = [1200,1500,2500,3500,4500]
plt.scatter(inc,loan)
```

```python
#pie chart
expenses = [20,20,15,30,15]
things = ['shopping','food','entertainment','parlour','gym']
plt.pie(expenses,labels=things,autopct="%1.1f")
plt.title("Monthly Expenses")
plt.show()
```
