```python
num = [10, 20, 30, 40, 50]
total = 0
for n in num:
    total += n
print("Sum of all elements:", total)
```

```python
nums = [5, 2, 9, 1, 7, 3]
minimum = nums[0]
maximum = nums[0]
for num in nums:
    if num < minimum:
        minimum = num
    if num > maximum:
        maximum = num
print("List:", nums)
print("Minimum:", minimum)
print("Maximum:", maximum)
```

```python
nums = [10, 20, 30, 40, 50]
reversed_list = []
for i in range(len(nums)-1, -1, -1):
    reversed_list.append(nums[i])
print("Reversed List:", reversed_list)
```

```python
nums = [2, 4, 6, 2, 8, 2, 4]
x = 4
count = 0
for num in nums:
    if num == x:
        count += 1
print("Element:", x)
print("Count:", count)
```

```python
nums = [1,2,3,4,5,6,7]
odd_numbers = []
for n in nums:
    if n % 2 != 0:
        odd_numbers.append(n)
print("Odd numbers:", odd_numbers)
```
