```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        fact=1
        for i in range(1,n+1):
            fact=fact*i
        return fact
factorial(5)
```

```python
def reverse(num):
    rev = 0
    while num!=0:
        rem = num%10
        rev = rev*10+rem
        num = num//10
    return rev
num = 572106
reverse(num)
```
